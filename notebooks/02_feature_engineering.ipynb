```python
import pandas as pd

df = pd.read_csv("../data/netflix_titles.csv")

df['country'] = df['country'].fillna("Unknown")
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year

df['genre_count'] = df['listed_in'].apply(lambda x: len(str(x).split(',')))
df['desc_length'] = df['description'].apply(lambda x: len(str(x)))
df['duration'] = df['duration'].str.extract(r'(\d+)').astype(float)

df['type'] = df['type'].map({"Movie":0, "TV Show":1})

df = df[['type','release_year','year_added','genre_count','desc_length','duration']]
df = df.dropna()

df.to_csv("../data/processed.csv", index=False)
```
